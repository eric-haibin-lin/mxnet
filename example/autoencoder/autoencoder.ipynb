```python
from matplotlib import pyplot as plt
import numpy as np
X = np.load('matrix_30_40_0.8_20_25_0.85_10_15_0.9_250_350_0.06.npy')
Y = np.load('prediction_430_128_4.npy')
Z = np.load('encode_430_128_4.npy')

plt.imshow(X, interpolation='nearest')
plt.show()
```

```python
plt.imshow(Y, interpolation='nearest')
plt.show()
```

```python
plt.imshow(Z, interpolation='nearest')
plt.show()
```

```python
def softmax(rows):
    row_max = np.max(rows, axis=1).reshape(-1, 1)
    e_x = np.exp(rows - row_max)
    return e_x / e_x.sum(axis=1).reshape(-1, 1)

c1 = softmax(Z[0:30, :])
c2 = softmax(Z[30:50, :])
c3 = softmax(Z[50:60, :])
c4 = softmax(Z[60:, :])

np.set_printoptions(suppress=True)
```

```python
plt.imshow(c1)
plt.show()
print(c1)
```

```python
plt.imshow(c2)
plt.show()
print(c2)
```

```python
plt.imshow(c3)
plt.show()
print(c3)
```

```python
plt.imshow(c4[:30, :])
plt.show()
print(c4[:30, :])
```

```python
plt.imshow(c4[:, :])
plt.show()
print(c4[:, :])
```
